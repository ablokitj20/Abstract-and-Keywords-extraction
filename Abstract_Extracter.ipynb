```python
import requests
from bs4 import BeautifulSoup
#libraries required to extract text
import time
```

```python
base_url = "https://www.mrs.org"
```

```python
fall_links=[]
spring_links=[]
#array topp hold the links of sympo
```

```python
for year in range(2011,2019):
    if(year<2013):
        fall_links.append(base_url + "/Fall" + str(year))
        spring_links.append(base_url + "/spring" + str(year))
        continue
    if(year<2016):
        fall_links.append(base_url + "/Fall" + str(year))
        spring_links.append(base_url + "/spring-" + str(year))
    else:
        fall_links.append(base_url + "/Fall-" + str(year)+"-symposium-sessions")
        spring_links.append(base_url + "/spring-" + str(year)+"-symposium-sessions")
```

```python
for i in range(3):
    s = "https://www.mrs.org/meetings-events/spring-meetings-exhibits/past-spring-meetings/"+ str(2019+i) +"-mrs-spring-meeting/call-for-papers"
    spring_links.append(s)
```

```python
fall_links.append("https://www.mrs.org/meetings-events/fall-meetings-exhibits/past-fall-meetings/2019-mrs-fall-meeting/call-for-papers")
fall_links.append("https://www.mrs.org/past-fall-meetings/2020-mrs-spring-and-fall-meeting/call-for-papers")
```

```python
def extract_topics(url,arr): #function to extract topics from the fall and spring links
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    sfTxtContent = soup.find_all(class_='sfTxtContent')
    for content in sfTxtContent:
        arr.append(content.text.strip())
```

```python
def _topics(url,arr): #function to extract topics from the fall and spring links after 2018
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    sfTxtContent = soup.find_all(class_='sfitemTitle sftitle')
    for content in sfTxtContent:
        arr.append(content.text.strip())
```

```python

def extract_links(url,arr): #function to extract links from the fall and spring links
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    for item in soup.find_all('li', class_='sfnewsListItem sflistitem'):
        link = item.find('a', {'href': True})
        if link:
            arr.append(base_url+link['href'])
```

```python
def _links(url,program_links): #function to extract links from the fall and spring links after 2018
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
#     program_links = []

    for li in soup.find_all('li', {'class': 'sfnewsListItem sflistitem'}):
        for a in li.find_all('a'):
            if a.text.strip() == 'Program':
                program_links.append(a['href'])
    
    for i in range(len(program_links)):
        program_links[i] = base_url + program_links[i]
```

```python
Fall_Topics=[]
Spring_Topics=[]
#each element will consist of all the fall and spring topics of each year
Fall_Links=[]
Spring_Links=[]
#each element will consist of all the fall and spring links of each year
```

```python
start_time = time.time()
i=0
for url in fall_links:
    if i<8:
        flag=[]
        extract_topics(url,flag)
        Fall_Topics.append(flag)
        flag=[]
        extract_links(url,flag)
        Fall_Links.append(flag)
    else:
        flag=[]
        _topics(url,flag)
        Fall_Topics.append(flag)
        flag=[]
        _links(url,flag)
        Fall_Links.append(flag)
    i+=1
end_time = time.time()
time_taken = end_time - start_time
print("Time taken:", time_taken, "seconds")
```

```python
start_time = time.time()
i=0
for url in spring_links:
    if i<8:
        flag=[]
        extract_topics(url,flag)
        Spring_Topics.append(flag)
        flag=[]
        extract_links(url,flag)
        Spring_Links.append(flag)
    else:
        flag=[]
        _topics(url,flag)
        Spring_Topics.append(flag)
        flag=[]
        _links(url,flag)
        Spring_Links.append(flag)
    i+=1
end_time = time.time()
time_taken = end_time - start_time
print("Time taken:", time_taken, "seconds")
```

#### Now we will go through each symposium and extract the abstracts

```python
# These Arrays will become 3D and will contain
# all the abstracts and their respecive titles
Fall_Abstracts=[]
Fall_Titles = []
Spring_Abstracts=[]
Spring_Titles = []
```

function to extract abstracts and titles

```python
def Extract_Info(url): 
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    abstracts = soup.find_all(class_='abstract')
    
    final_abstracts = []
    final_titles = []
    title=[]
    
    for abstract in abstracts:
        text = str(abstract)
        index = text.find('<b>') + len('<b>')
        index0 = text.find('</b>')
        index1 = text.find('<div>') + len('<div>')
        index2 = text.find('</div>', index1)
        index3 = text.find('<div>', index2) + len('<div>')
        index4 = text.find('</div>', index3)
        abstract_text = abstract.text.strip() 
        if len(abstract_text) > 100:
            final_abstracts.append(abstract_text)
            final_titles.append(text[index:index0] + " " + text[index3:index4])
    
    return final_abstracts, final_titles
```

Nested loop to go through each abstract from each Symposium from each year

```python
start_time = time.time()
for links in Fall_Links:
    flag1 = []
    flag2 = []
    
    for url in links:
        final_abstracts, final_titles = Extract_Info(url)
        flag1.append(final_abstracts)
        flag2.append(final_titles)
    
    Fall_Abstracts.append(flag1)
    Fall_Titles.append(flag2)

for links in Spring_Links:
    flag1 = []
    flag2 = []
    
    for url in links:
        final_abstracts, final_titles = Extract_Info(url)
        flag1.append(final_abstracts)
        flag2.append(final_titles)
    
    Spring_Abstracts.append(flag1)
    Spring_Titles.append(flag2)
end_time = time.time()
time_taken = (end_time - start_time)/60
print(f"time taken: {time_taken:.2f} mins")
```

```python
#function to strip the required abstract
def only(texts):
    start_index = texts.find('Show Abstract') + 13
    return (texts[start_index:].strip())
    
```

```python
for i in range(len(Fall_Abstracts)):
    for j in range(len(Fall_Abstracts[i])):
        for k in range(len(Fall_Abstracts[i][j])):
            Fall_Abstracts[i][j][k] = only(Fall_Abstracts[i][j][k])
```

```python
for i in range(len(Spring_Abstracts)):
    for j in range(len(Spring_Abstracts[i])):
        for k in range(len(Spring_Abstracts[i][j])):
            Spring_Abstracts[i][j][k] = only(Spring_Abstracts[i][j][k])
```

```python
#libraries to save the abstracts to a pdf
import xml.sax.saxutils
from reportlab.lib.pagesizes import letter
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import SimpleDocTemplate, Paragraph
from reportlab.lib.units import inch
from IPython.display import FileLink
```

```python
start_time = time.time()
doc = SimpleDocTemplate("Fall_Abstract.pdf", pagesize=letter,
                        rightMargin=72, leftMargin=72,
                        topMargin=72, bottomMargin=18)

styles = getSampleStyleSheet()

# Defining style for heading
heading_style = styles["Heading1"]

# Defining style for sub-heading
sub_heading_style = styles["Heading2"]

# Defining style for regular text
normal_style = styles["Normal"]
normal_style.fontName = "Helvetica"
title_style = styles["Heading4"]
title_style.fontName = "Helvetica-Bold"

# Creating list to hold all elements to be added to the PDF
elements = []

# Loop through each element in the 3D array and add it to the PDF
for i in range(len(Fall_Abstracts)):
    # Adding heading for ith symposium
    heading = f"201{i+1}-Symposiums"
    elements.append(Paragraph(heading, heading_style))
    
    for j in range(len(Fall_Abstracts[i])):
        # Adding sub-heading for jth presentation
        sub_heading = Fall_Topics[i][j]
        elements.append(Paragraph(sub_heading, sub_heading_style))
        
        for k in range(len(Fall_Abstracts[i][j])):
            title = Fall_Titles[i][j][k] + '\n'
            elements.append(Paragraph(title, title_style))
            abstract_text = Fall_Abstracts[i][j][k]
            abstract_text = xml.sax.saxutils.escape(abstract_text)
            elements.append(Paragraph(abstract_text, normal_style))
            
            # Adding newline gap at end of abstract
            elements.append(Paragraph("<br/>", normal_style))

# Adding all elements to the PDF
doc.build(elements)
end_time = time.time()
time_taken = (end_time - start_time)/60
print(f"time taken: {time_taken:.2f} mins")
```
